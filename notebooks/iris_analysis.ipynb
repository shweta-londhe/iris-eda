```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='species')
plt.title("Species Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue="species")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df.boxplot()
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue="species")
plt.show()
```

```python
print(" Iris Dataset  Key Insights Summary\n")

print("1 Setosa species:")
print("- Very small petal length & width")
print("- Clearly separated from the other species in all plots\n")

print("2 Versicolor species:")
print("- Medium petal size")
print("- Some overlap with Virginica\n")

print("3 Virginica species:")
print("- Largest petal measurements")
print("- Clearly separated from Setosa\n")

print(" Overall Conclusion:")
print("- Petal measurements (length & width) are the strongest features to distinguish species.")
print("- Setosa is easiest to identify.")
print("- Versicolor & Virginica show partial overlap but can still be separated with good accuracy.")
```
